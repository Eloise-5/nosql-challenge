## Notebook Set Up

```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
print(db.list_collection_names())
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

# Use count_documents to display the number of documents in the result
print("Number of establishments with a hygiene score of 20: ", establishments.count_documents(query))

# Display the first document in the results using pprint
print("First result:")
results = establishments.find(query)
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
result_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(result_df))

# Display the first 10 rows of the DataFrame
result_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query2 = establishments.find({"LocalAuthorityName": {'$regex': 'London'}, "$or":[ {"RatingValue":4}, {"RatingValue":5}]})

# Use count_documents to display the number of documents in the result
print("Number of establishments in London with a Rating Value greater than or equal to 4 ", establishments.count_documents(query2))

# Display the first document in the results using pprint
print("First result:")
results2 = establishments.find(query2)
pprint(results2[0])
```

```python
# Convert the result to a Pandas DataFrame
result_df2 = pd.DataFrame(results2)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(result_df2))

# Display the first 10 rows of the DataFrame
result_df2.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 0.08384000
longitude = 51.49014200

query3 = {'$match': {'RatingValue': 5},
                          'geocode.latitude': {'$lte': (latitude + degree_search), '$lte': (latitude - degree_search)},
                          'geocode.longitude': {'$lte': (longitude + degree_search), '$lte': (longitude - degree_search)}
                         }

sort3 = {'$sort': {'scores.Hygiene': 1}} 
limit3 = 5

results3= list(establishments.find(query3).sort(sort3).limit(limit3))

# Print the results
pprint(results3)
```

```python
# Convert result to Pandas DataFrame
neighbour_highscore = pd.DataFrame(results3)
neighbour_highscore
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

match_4 = {'$match': {'scores.Hygiene': "0"}}

group_4 = {'$group': {'_id': "$LocalAuthorityName"}}

sort_4 = {'$sort': { 'count': -1, '_id': 1 }}

pipeline4 = [match_4, group_4, sort_4]
results4 = list(establishments.aggregate(pipeline4))
                      
# Print the number of documents in the result
print("Number of classifications in result: ", len(results4))
                      
# Print the first 10 results
pprint(results4[0:10])
```

```python
# Convert the result to a Pandas DataFrame
results4_df = pd.DataFrame(results4)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(results4_df))

# Display the first 10 rows of the DataFrame
results4_df.head(10)
```
